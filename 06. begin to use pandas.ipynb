```python
#convert NiceCX to a networkx object in order to retrieve all the nodes attributes
mynetwork_NetworkX = mynetwork_from_server.to_networkx()

#get the number of nodes in the network
number_of_nodes = mynetwork_NetworkX.number_of_nodes()
print(number_of_nodes)

#display the detailed information of the nodes
#node_list_NetworkX = list(mynetwork_NetworkX.nodes(data=True))
#for i in range(len(node_list_NetworkX)):
#    print(node_list_NetworkX[i], "\n")
```

```python
#create an empty dataframe 
my_dataframe = pd.DataFrame()

#using a loop to convert all nodes to Pandas DataFrame one by one
for i in range(number_of_nodes):
    #get each node's name
    node_name = list(mynetwork_NetworkX.nodes(data=True))[i][0]
    #get each node's attributes with values, this returns a data dict
    node_attributes = list(mynetwork_NetworkX.nodes(data=True))[i][1]

    #add node name to the dictionary, 'Name' is at the end of the data dict
    node_attributes.update({'Name':node_name}) 

    #creating a dataFrame from the data dictionary. This dataframe only has one row.
    df = pd.DataFrame(node_attributes, index=[i])
 
    #add the one row dataframe to my_dataframe. 
    #Must include the argument 'sort=True', otherwise cannot include "Name" in the dataFrame column list
    my_dataframe = pd.concat([my_dataframe, df], ignore_index=True, sort=True)

#display the dataframe
my_dataframe
```

```python
#reorder the columns, keep the "Name" attribute in the first column
first_column = my_dataframe['Name']
my_dataframe.drop(labels=['Name'], axis=1, inplace=True)
my_dataframe.insert(0, 'Name', first_column)
my_dataframe
```
