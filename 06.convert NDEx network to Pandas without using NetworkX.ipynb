```python
import ndex2.client as nc
import io
import json
from time import sleep

from ndex2.nice_cx_network import NiceCXNetwork
import ndex2

import pandas as pd 
import numpy as np

#create an NDEx client object to access my accouont on the NDEx server
my_account="Jeffreyc"
my_password="Frenchfri3"

my_ndex=nc.Ndex2("http://public.ndexbio.org", my_account, my_password)
my_ndex.update_status()
```

```python
#get NiceCX object from server 
mynetwork_from_server = ndex2.create_nice_cx_from_server(server='http://public.ndexbio.org',  uuid='3724db72-2022-11e9-bb6a-0ac135e8bacf')

#get number of nodes in the network and number of attributes for a node
mynetwork_from_server.print_summary()
```

```python
# create an empty dataframe in order to convert the network.
my_dataframe = pd.DataFrame()

#using a loop to retrieve all the nodes one by one
for id, node in mynetwork_from_server.get_nodes() :
    
    #get this node's attributes, attribute_list is an array, each index of the array is a dictionary
    attribute_list = mynetwork_from_server.get_node_attributes(id)
    number_of_attribute = len(attribute_list)
    
    # create 2 arrays for each node: one for attribute name and the other one for attribute value
    list_name = []
    list_value =[] 
    
    # using a loop to add the node's attribute names and values to the two arrays
    # attribute_list is an array, each index of the array is a dictionary, 
    # attribute_list[i].values() returns an array shown as [node id, attribute name, attribute value]
    for i in range (number_of_attribute):
        list_name.append(list(attribute_list[i].values())[1])
        list_value.append(list(attribute_list[i].values())[2])
        
    # create a one row dataframe using the attribure arrays, convert this node to a one row dataframe
    # pay attention to the syntax !!!
    df = pd.DataFrame([list_value],columns=list_name)
    
    #add the one row dataframe to my_dataframe.
    my_dataframe = pd.concat([my_dataframe, df], ignore_index=True, sort=True)
    
my_dataframe
```

```python
# replace all NaNs with an empty string
my_dataframe = my_dataframe.fillna('')

# add a column "Name" with empty string to the left of the dataframe
my_dataframe.insert(0,"Name",'')

my_dataframe
```

```python
#convert the Pandas DataFrame to NiceCX object and upload to my account
#compare the uploaded network with the original network, should see that all edges are changed
my_new_NiceCX = ndex2.create_nice_cx_from_pandas(my_dataframe)

my_new_NiceCX.upload_to("http://public.ndexbio.org", my_account, my_password)
```
